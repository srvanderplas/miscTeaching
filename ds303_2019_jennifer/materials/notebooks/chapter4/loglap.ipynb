```python
import numpy as np
import pylab as plt
%matplotlib inline
```

Generate some data with two classes

```python
x = np.random.randn(50,2)
x[:25,:] += 1.5
x[25:,:] += -1.5
t = np.zeros((50,1))
t[25:] += 1
```

Plot the data

```python
plt.figure()
styles = ['ro','bo']
classes = np.unique(t)
for i,classn in enumerate(classes):
    pos = np.where(t==classn)[0]
    plt.plot(x[pos,0],x[pos,1],styles[i])
```

Use Newton-Raphson to find the $\mathbf{w}$ that maximises the likelihood

```python
sig_sq = 0.5 # prior variance for w (assuming mean zero)
```

```python
def grad_function(w,x,t,sig_sq):
    P = 1.0/(1.0 + np.exp(-np.dot(x,w)))
    return (-1.0/sig_sq)*w + np.dot(x.T,t-P)
```

```python
def hess_function(w,x,t,sig_sq):
    P = 1.0/(1.0 + np.exp(-np.dot(x,w)))
    P = P.flatten() # required for the diagonalisation
    return -(1.0/sig_sq)*np.eye(len(w)) - np.dot(x.T,np.dot(np.diag(P*(1-P)),x))
```

```python
w = np.zeros((2,1)) # Initial guess

all_w = []
all_w.append(w.flatten())
for it in range(10):
    w = w - np.dot(np.linalg.inv(hess_function(w,x,t,sig_sq)),grad_function(w,x,t,sig_sq))
    all_w.append(w.flatten())
```

Plot the convergence of the w values

```python
plt.figure()
all_w = np.array(all_w)
for i in range(2):
    plt.plot(all_w[:,i])
```

Define the Gaussian approximation

```python
g_mean = w
H = hess_function(g_mean,x,t,sig_sq)
g_cov = np.linalg.inv(-H)
```

Sample some decision boundaries from this approximation and plot them

```python
plt.figure()
for i,classn in enumerate(classes):
    pos = np.where(t==classn)[0]
    plt.plot(x[pos,0],x[pos,1],styles[i])

w_samps = np.random.multivariate_normal(g_mean.flatten(),g_cov,20)
for w_sample in w_samps:
#     Plot the line corresponding to w.Tx = 0
    xv = np.array([-5,5])
    yv = (-w_sample[1]/w_sample[0])*xv
    plt.plot(xv,yv)
plt.xlim([-5,5])
plt.ylim([-5,5])
    
```

Plot the decision boundary. We will actually plot probability contours by making a big grid. To get these values, we have to take an expectation over the approximate posterior (the Laplacian) which we in turn approximate by sampling

```python
gridX,gridY = np.meshgrid(np.arange(-5,5,0.1),np.arange(-5,5,0.1))
```

```python
n_samples = 1000
w_samples = np.random.multivariate_normal(g_mean.flatten(),g_cov,n_samples)
```

Note the following code is vectorised to make it faster (needs no explicit loops)

```python
nr,nc = gridX.shape
gx = np.reshape(gridX,(nr*nc,1))
gy = np.reshape(gridY,(nr*nc,1))
g = np.hstack((gx,gy))
P = (1.0/(1+np.exp(np.dot(g,w_samples.T)))).mean(axis=1)
P = np.reshape(P,(nr,nc))
```

```python
plt.figure()
for i,classn in enumerate(classes):
    pos = np.where(t==classn)[0]
    plt.plot(x[pos,0],x[pos,1],styles[i])
plt.contour(gridX,gridY,P)
```
